```python
# The Building Blocks

year = '2022'
url_link = 'https://www.basketball-reference.com/leagues/NBA_{}_per_game.html'
```

```python
# Combining the URL + year strings together

url = url_link.format(year)
url
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pylab as plt
%matplotlib inline
sns.set(color_codes=True)
plt.style.use('ggplot')
```

```python
# using pd.read_html() function to read HTML webpage

df = pd.read_html(url, header = 0)
df
```

```python
# to check how many table are there in a webpage

len(df)
```

```python
# Selecting the first table in a webpage

df[0]
```

```python
df2022 = df[0]
```

```python
# Data cleaning

# We can see that the table header is presented multiple times in several rows.

df2022[df2022.Age == 'Age']
```

```python
len(df2022[df2022.Age == 'Age'])
```

```python
df = df2022.drop(df2022[df2022.Age == 'Age'].index)
     
```

```python
df.shape
```

```python
# Understanding of all variables names in our data

df.columns
```

```python
# Checking for the datatype that our dataset is consist of

df.dtypes
```

```python
# Checking for missing values in our that before analysis

print(df.isnull().sum())
```

```python
# Replace null values with o in the data

df = df.fillna(0)
```

```python
# Remove irrelant column of Rk

df = df.drop(['Rk'], axis=1)
```

```python
df.head()
```

```python
# Write to csv file

df.to_csv('nba2022.csv', index=False)
```

```python
# Use condition selection of subset data to perform analysis and comparison

### Which player scored the most points per game?

df[df.PTS == df.PTS.max()]
```

```python
### What team a player from?

PlayerMaxPoints = df[df.PTS == df.PTS.max()]
PlayerMaxPoints.Tm
```

```python
## Which Position is a player playing in a team?

PlayerMaxPoints.Pos
```

```python
### How many game did a player played in a season?

PlayerMaxPoints.G
```

```python
### Which player scored more than 20 Points(PTS) per game?

df[df.PTS > 20]
```

```python
### Which player had the highest 3-Point field goals per game?

df[df['3P'] == df['3P'].max()]
```

```python
### Which player had the most assist(AST) per game?

df[df['AST'] == df['AST'].max()]
```

```python
### Used GROUPBY function to determine the player who scored highest point in a Los Angeles Lakers 

LAL = df.groupby('Tm').get_group('LAL')
LAL
```

```python
### Out of 5 position, which position scores the most points?
#### first group players by positions 

df.groupby('Pos').PTS.describe()
```

```python
#### Only select the 5 traditional positions

positions = ['C', 'PF', 'SF', 'PG','SG']
POS = df[df['Pos'].isin(positions)]
POS
```

```python
#### Take a look at descriptive statistic of all positions

POS.groupby('Pos').PTS.describe()
```

```python
### Create subset dataframe of Position(Pos) and Points(PTS) using only selected positions values

PTS = df[['Pos', 'PTS']]
positions = ['C', 'PF', 'SF', 'PG','SG']
PTS = PTS[PTS['Pos'].isin(positions)]
PTS
```

```python
### Checking for the distribution of data among 5 selected positions

PTS['PTS'].hist(by=PTS['Pos'], layout=(1,5), figsize=(16,2))
```

```python
### Understand the distribution better by applying boxplot

sns.boxplot(x='Pos', y='PTS', data = PTS)
```

```python
### Compute the correlation matrix of the data

corr = df.corr()
corr
```

```python
sns.heatmap(corr)
```

```python
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True
with sns.axes_style("white"):
    f, ax = plt.subplots(figsize=(7,5))
    ax = sns.heatmap(corr, mask=mask, vmax=1, square=True)
```
