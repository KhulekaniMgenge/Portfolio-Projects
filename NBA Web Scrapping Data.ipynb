```python
# The Building Blocks

year = '2022'
url_link = 'https://www.basketball-reference.com/leagues/NBA_{}_per_game.html'
```

```python
# Combining the URL + year strings together

url = url_link.format(year)
url
```

```python
import pandas as pd
```

```python
# using pd.read_html() function to read HTML webpage

df = pd.read_html(url, header = 0)
df
```

```python
# to check how many table are there in a webpage

len(df)
```

```python
# Selecting the first table in a webpage

df[0]
```

```python
df2022 = df[0]
```

```python
# Data cleaning

# We can see that the table header is presented multiple times in several rows.

df2022[df2022.Age == 'Age']
```

```python
len(df2022[df2022.Age == 'Age'])
```

```python


df = df2022.drop(df2022[df2022.Age == 'Age'].index)
     
```

```python
df.shape
```

```python
import seaborn as sns
```

```python
# making the histogram

sns.distplot(df.PTS, 
             kde=False)
```

```python
# change bar line color

sns.distplot(df.PTS, 
             kde=False,
             hist_kws=dict(edgecolor="black", linewidth=2))
```

```python
# Change bar fill colors

sns.distplot(df.PTS,
             kde=False,
             hist_kws=dict(edgecolor="black", linewidth=2),
             color='#00BFC4')
```
